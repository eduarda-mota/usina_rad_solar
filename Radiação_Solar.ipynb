```python
# Importação das bibliotecas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Leitura, Visualização e Organização dos Dados

```python
# Leitura dos dados

rad_global = pd.read_excel('/content/drive/MyDrive/Meteorologia Física II/INMET_N_AM_A128_BARCELOS_01-01-2023_A_31-12-2023.CSV.xlsx')
```

```python
rad_global.head()
```

```python
rad_global.dtypes
```

```python
# Substituindo NaN por zero (na coluna radiacao_global)

rad_global["radiacao_global"] = rad_global["radiacao_global"].fillna(0)
```

```python
rad_global.head(10)
```

```python
# Dimensão dos Dados

rad_global.shape
```

```python
# Campos nulos (NaN)

rad_global.isnull().sum()
```

```python
# Valores negativos

rad_global.loc[rad_global.radiacao_global < 0, :].count()
```

## Valores Brutos da Radiação Global

```python
# Gráfico de Linha

plt.figure(figsize=(14,6))
plt.plot(rad_global['Data'], rad_global['radiacao_global'])
plt.title('Radiação Global - Barcelos (AM)')
plt.ylabel('Radiação Global (Kj/m²)')
plt.xlabel('Data');
#plt.xticks(rotation=45, ha='right');
```

## Médias Diárias da Radiação Global

```python
# Selecionando as linhas em que a radiação global é diferente de zero

rad_global_filtrado = rad_global[rad_global['radiacao_global'] != 0]
```

```python
# Excluindo a coluna 'Hora UTC'

rad_global_filtrado = rad_global_filtrado.drop('Hora UTC', axis=1)
```

```python
# Colocando a coluna 'Data' como index

rad_global_filtrado.set_index('Data', inplace=True)
```

```python
rad_global_filtrado.head(4)
```

```python
# Cálculo das médias diárias

rad_global_diario = rad_global_filtrado["radiacao_global"].resample('D').mean()
```

```python
rad_global_diario.head(5)
```

```python
print(type(rad_global_diario))
```

```python
rad_global_diario = rad_global_diario.to_frame(name='radiacao_global')
rad_global_diario.reset_index(inplace=True)
```

```python
rad_global_diario
```

```python
# Gráfico de Linha

plt.figure(figsize=(14,6))
plt.plot(rad_global_diario['Data'], rad_global_diario['radiacao_global'])
plt.title('Médias Diárias da Radiação Global - Barcelos (AM)')
plt.ylabel('Média Diária (Kj/m²)')
plt.xlabel('Data');
#plt.xticks(rotation=45, ha='right');
```

## Médias Mensais da Radiação Global

```python
# Cálculo das médias mensais

rad_global_mensal = rad_global_filtrado["radiacao_global"].resample('M').mean()
rad_global_mensal = rad_global_mensal.to_frame(name='radiacao_global')
rad_global_mensal.reset_index(inplace=True)
rad_global_mensal
```

```python
# Gráfico de Linha

plt.figure(figsize=(15,6))
plt.plot(rad_global_mensal['Data'], rad_global_mensal['radiacao_global'], color='purple')
plt.title('Médias Mensais da Radiação Global - Barcelos (AM)')
plt.ylabel('Média Mensal (Kj/m²)')
plt.xlabel('Data');
#plt.xticks(rotation=45, ha='right');
```

```python
# Gráfico de Colunas

plt.figure(figsize=(14,6))
sns.barplot(data=rad_global_mensal, x='Data', y='radiacao_global', color='purple')
plt.title('Médias Mensais da Radiação Global - Barcelos (AM)')
plt.ylabel('Média Mensal (Kj/m²)')
plt.xlabel('Data');
#plt.xticks(rotation=45, ha='right');
```

## Média Anual da Radiação Global

```python
# Cálculo da média anual

rad_global_anual = rad_global_filtrado["radiacao_global"].resample('Y').mean()
rad_global_anual = rad_global_anual.to_frame(name='radiacao_global')
rad_global_anual.reset_index(inplace=True)
rad_global_anual
```
